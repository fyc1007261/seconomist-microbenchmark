```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_server = pd.read_csv('crypto-server.csv', header=None, names=['size', 'hash', 'sign', 'verify'])
```

```python
df_server
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8,3))
```

```python
axs[0].plot(df_server['size'], df_server['hash'], '-', label='hash')
axs[0].plot(df_server['size'], df_server['sign'], '--', label='sign')
axs[0].plot(df_server['size'], df_server['verify'], '-.', label='verify')
axs[0].legend()

fig
```
